### Esercizio  2.1
Viene calcolato il valore dell'integrale: $I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$ con un metodo Monte Carlo, campionando prima una distribuzione uniforme, poi un'apposita distribuzione per verificare l'utilità dell'importance sampling. <br>
Per l'importance sampling la distribuzione utilizzata è d(x)=-2x+2, e come si vede dai grafici riportati in seguito, a parità di punti campionati assicura un risultato più preciso.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
c1,c2,c3 = np.loadtxt("data/output.dat", usecols=(0,1,2), unpack=True)
c4,c5,c6 = np.loadtxt("data/output2.dat", usecols=(0,1,2), unpack=True)
plt.figure(figsize=(6.4*1.5, 4.8*1.5))
plt.subplot(2,1,1)
plt.errorbar(c1, c2, yerr=c3)
plt.hlines(1,0, 100)
plt.ylabel("Integral value (uniform sampling)")
plt.grid(True)
plt.subplot(2,1,2)
plt.errorbar(c4, c5, yerr=c6)
plt.hlines(1,0, 100)
plt.xlabel("number of blocks (1000 throws each)")
plt.ylabel("Integral value (sample through d(x)=-2x+2)")
plt.grid(True)
plt.show()
```
