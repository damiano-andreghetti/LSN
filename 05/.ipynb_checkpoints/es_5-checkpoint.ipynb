# Lezione di laboratorio 5
In questa lezione abbiamo implementato e utilizzato l'algoritmo di Metropolis. <br>
Il fine è di campionare il modulo quadro di funzioni d'onda dell'atomo  di idrogeno per poi calcolare il valor medio della posizione (ovvero$\langle \sqrt{r^2} \rangle$). <br>
Le funzioni d'onda campionate hanno la forma:
1. 1s: $$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
2. 2p: $$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$ <br> <br>
Per queste i valori medi della posizione sono noti e sono rispettivamente $\frac{3}{2}a_0$ e $5a_0$ 

### Esercizio 5.1
Usando il metodo di Metropolis sono state campionate le distribuzioni $\mid \psi_{1,0,0} \mid^2$ e $\mid \psi_{2,1,0} \mid^2$.
A seguito sono riportati i grafici di $\langle \sqrt{r^2} \rangle$ in funzione del numero di blocchi in cui sono state divise le simulazioni.<br>
Per $\mid \psi_{1,0,0} \mid^2$ con probabilità di transizione uniforme:

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt;
a1, a2, a3 = np.loadtxt("data/output.dat", usecols=(0,1,2), unpack=True)
plt.errorbar(a1, a2, yerr=a3)
plt.ylabel("<|r|> in unità del raggio di Bohr")
plt.hlines(1.5, 0, 100)
plt.grid(True)
plt.show()
```

Per $\mid \psi_{2,1,0} \mid^2$ con probabilità di transizione uniforme.

```python
b1, b2, b3 = np.loadtxt("data/output2.dat", usecols=(0,1,2), unpack=True)
plt.errorbar(b1, b2, yerr=b3)
plt.ylabel("<|r|> in unità del raggio di Bohr")
plt.hlines(5, 0, 100)
plt.grid(True)
plt.show()
```

A seguito invece è riportato lo stesso risultato con campionamento di $\mid \psi_{2,1,0} \mid^2$ con probabilità di transizione gaussiana.

```python
c1, c2, c3 = np.loadtxt("data/output3.dat", usecols=(0,1,2), unpack=True)
plt.errorbar(c1, c2, yerr=c3)
plt.ylabel("<|r|> in unità del raggio di Bohr")
plt.hlines(5, 0, 100)
plt.grid(True)
plt.show()
```
