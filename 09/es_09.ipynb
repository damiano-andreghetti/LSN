# Lezione di laboratorio 9
In questa lezione è stato implementato in c++ un algoritmo genetico per risolvere un problema TSP (Travelling Salesman Problem). <br>
Un algorimto genetico è un metodo euristico per risolvere una certa categoria di problemi. <br>
Supponendo di poter codificare una possibile soluzione all'interno di una sequenza, e di poter valutare con una funzione costo una soluzione, allora vengono generate un certo numero N di sequenze (o soluzioni random). <br>
Queste sequenze sono poi incrociate fra di loro, selezionando delle coppie in base alla funzione costo (sarà più probabile che vengano selezionate per l'"accoppiamento" le soluzioni migliori). <br>
Una volta ottenuta una seconda generazione di N sequenze il processo viene reiterato. <br>

Un TSP è un problema definito da una serie di punti attraverso cui deve passare un percorso (una sola volta per ogni punto). Si cerca il percorso più breve. Avendo anche solo 30 punti il numero di possibili percorsi è 30! e quindi risulta un problema difficile da risolvere con metodi numerici "standard".

### Esercizio 8.1a
30 punti casuali sono generati su una circonferenza (r=1), e si cerca il percorso più breve che passa da tutti i punti una sola volta.
Ogni possibile percorso è rappresentabile da una sequenza di numeri ognuno rappresentate una città.
Sono generati 900 percorsi random, ordinati in base alla lunghezza del percorso e incrociati utilizzando come probabilità di selezione $j=N_p r^2$ dove N è il numero di elementi della popolazione e r un numero casuale uniforme.<br>
Le sequenze di città sono incrociate ed inoltre subiscono mutazioni (in modo da assicurarsi che esplorino più configurazioni possibili).<br>
In tutto vengono create 100 generazioni di soluzioni.
Di seguito sono riportati due grafici:
- funzione costo del miglior individuo di ogni generazione (in funzione del numero della generazione) 
- costo medio della migliore metà della popolazione di ogni generazione (in funzione del numero della generazione)

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
#from matplotlib.pyplot import figure
b1 = np.loadtxt("data/best_stats1.dat")
b2 = np.arange(b1.size)
a1 = np.loadtxt("data/average_stats1.dat")
a2 = np.arange(a1.size)
plt.title("cost of the best individual")
plt.plot(b2,b1)
plt.show()
```

```python
plt.title("average cost of best half of the population")
plt.plot(a2,a1)
plt.show()
```

#### Miglior percorso trovato alla 100esima generazione

```python
bp = np.genfromtxt("data/best_individual1.dat", usecols=(0), dtype=np.int8, unpack=True, skip_footer=31)
cities_x, cities_y = np.genfromtxt("data/best_individual1.dat", usecols=(0,1), unpack=True, skip_header=31)
citx=cities_x[bp]
city=cities_y[bp]
plt.figure(figsize=(6, 6))
plt.plot(citx,city,"-o")
plt.show()
```
