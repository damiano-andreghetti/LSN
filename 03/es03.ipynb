# Lezione di laboratorio 3
Questa lezione è incentrata sulle applicazioni in economia del calcolo stocastico.
Ci si propone di calcolare call-option e put-option per opzioni di tipo Europeo (obbligazioni vendute a prezzo e data prestabiliti). <br>
Questo è stato fatto in due modi: campionando direttamente la funzione S(T) al tempo T desiderato (3.1); <br>
oppure calcolando reiterativamente i passi di un GBM (Generic Brownian Motion) fino al tempo T desiderato. <br>
In entrambi i casi i parametri utilizzati sono: <br>
- prezzo del bene: $t=0$: $S(0)=100$
- tempo di acquisto/vendita: $T=1$
- strike price: $K=100$
- tasso d'interesse senza rischi: $r=0.1$
- volatilità $\sigma=0.25$

<br>
Le rette orizzontali mostrano la soluzione analitica trovata con il metodo di Black-Scholes.

### Esercizio 3.1
Sapendo che la funzione prezzo del bene S(T) ha la forma:
    $S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$
    dove $W(t)\sim\mathcal{N}(0,t)$ <br>
e che il profitto è dato da p=max(0, S-K) per una call option e da p=max(0, K-S) per una put option; <br>
allora il profitto attuale sarà dato da $p\exp[-rT]$, ovvero rimuovendo il valore degli interessi che avrebbe fornito una banca depositando la somma invece che investendola.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
a1,a2,a3 = np.loadtxt("data/output.dat", usecols=(0,1,2), unpack=True)
b1,b2,b3 = np.loadtxt("data/output2.dat", usecols=(0,1,2), unpack=True)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.errorbar(a1,a2,yerr=a3)
plt.hlines(14.975790778311286, 0, 100)
plt.ylabel("price")
plt.xlabel("number of blocks(10000 throws each)")
plt.title("call option pricing through direct asset price sampling")
plt.grid()
plt.subplot(1,2,2)
plt.errorbar(b1,b2,yerr=b3)
plt.hlines(5.4595325819072364, 0, 100)
plt.ylabel("price")
plt.xlabel("number of blocks(10000 throws each)")
plt.title("put option pricing through direct asset price sampling")
plt.grid()
plt.show()
```

### Esercizio 3.2
Campionando gli step di un GBM ci si aspetta di giungere allo stesso risultato. <br>
L'intervallo di tempo tra 0 e T viene diviso in 100 sottointervalli (step).
Il procedimento è ricorsivo, infatti:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    con $Z_i \sim \mathcal{N}(0,1)$
Anche in questo caso viene calcolato il profitto e corretto rispetto al tasso d'interesse a rischio zero dato da una banca (r).

```python
c1,c2,c3 = np.loadtxt("data/output3.dat", usecols=(0,1,2), unpack=True)
d1,d2,d3 = np.loadtxt("data/output4.dat", usecols=(0,1,2), unpack=True)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.errorbar(c1,c2,yerr=c3)
plt.hlines(14.975790778311286, 0, 100)
plt.ylabel("price")
plt.xlabel("number of blocks(10000 throws each)")
plt.title("call option: asset price sampling through 100 intervals")
plt.grid()
plt.subplot(1,2,2)
plt.errorbar(d1,d2,yerr=d3)
plt.hlines(5.4595325819072364, 0, 100)
plt.ylabel("price")
plt.xlabel("number of blocks(10000 throws each)")
plt.title("put option: asset price sampling through 100 intervals")
plt.grid()
plt.show()
```
