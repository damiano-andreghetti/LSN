### Esercizio 7.1a
Una volta esportati i valori istantanei di U/N e P, si è utilizzato il seguente script python per costruiure la funzione di autocorrelazione per entrambe le siere di dati (energia interna per particella e pressione). <br>
È stato calcolato poi il fit con una funzione esponenziale.<br>
La funzione di autocorrelazione di una serie di dati $\{ y_k \}$ è definita come:
$$R(t)=\frac{\langle y_k y_{k+t} \rangle_k - \langle y_k \rangle_k \langle y_{k+t} \rangle_k}{\sigma_y^2}$$
<br>
Per U/N:

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
def func(x, tc):
    return  np.exp(-abs(x)/tc)


a1,a2 = np.loadtxt("data/insta.dat", usecols=(0,1),unpack=True)
sigmaUN = np.std(a1)
sigmaP = np.std(a2)
autoUN = np.zeros(400)
for tau in range(0, 400):
    acc=0
    c1=0
    c2=0
    c3=0
    for k in range (0,len(a1)-tau):
        c1+=a1[k]*a1[k+tau]
        c2+=a1[k]
        c3+=a1[k+tau]
        acc+=1
    c1/=acc
    c2/=acc
    c3/=acc
    autoUN[tau]=(c1-(c2*c3))/(sigmaUN*sigmaUN)
N=np.arange(400)
popt, pcov = curve_fit(func, N, autoUN)
plt.figure(figsize=(12.8,9.6))
plt.plot(N, autoUN)
plt.plot(N, func(N, *popt))
plt.legend(("instantaneous U/N autocorrelation", "exponential fit"))
print("coefficiente fit: ", str(popt))
plt.grid()
plt.show()
```

Per la pressione:

```python
a1,a2 = np.loadtxt("data/insta.dat", usecols=(0,1),unpack=True)
sigmaUN = np.std(a1)
sigmaP = np.std(a2)
autoP = np.zeros(400)
for tau in range(0, 400):
    acc=0
    c1=0
    c2=0
    c3=0
    for k in range (0,len(a2)-tau):
        c1+=a2[k]*a2[k+tau]
        c2+=a2[k]
        c3+=a2[k+tau]
        acc+=1
    c1/=acc
    c2/=acc
    c3/=acc
    autoP[tau]=(c1-(c2*c3))/(sigmaP*sigmaP)
N=np.arange(400)
popt, pcov = curve_fit(func, N, autoP)
plt.figure(figsize=(12.8,9.6))
plt.plot(N, autoP)
plt.plot(N, func(N, *popt))
plt.legend(("instantaneous P autocorrelation", "exponential fit"))
print("coefficiente fit: ", str(popt))
plt.grid()
plt.show()
```

# Esercizio 7.1b
Utilizzando il set dei valori istantanei di energia interna per particella con il seguente script python si calcola l'incertezza sui valori medi per varie dimensioni dei blocchi in cui sono raccolti i dati. <br>
La lunghezza dei blocchi L varia da 10 a 5000 valori per blocco, si può notare come l'incertezza cali rapidamente all'aumentare della dimensione dei blocchi fino a 1000 valori, per poi decrescere più lentamente

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import math
a1,a2 = np.loadtxt("data/insta.dat", usecols=(0,1),unpack=True)
M = len(a1)
L=10
c = 0
res = np.zeros(500)
while L < 5000:
    N=int(M/L)
    sum_prog = 0
    sum_prog2 = 0
    av = np.zeros(N)
    avq = np.zeros(N)
    for n in range (N):
        for i in range (L):
            av[n]+=a1[i]
            avq[n]+=a1[i]*a1[i]
        av[n]/=L
        avq[n]/=L
    sum_prog = np.sum(av)
    sum_prog2 = np.sum(avq)
    sum_prog/=N
    sum_prog2/=N
    res[c] = math.sqrt(abs(sum_prog2-sum_prog**2)/N)/L
    L+=10
    c+=1
x = np.arange(len(res))
x*=10
plt.plot(x, res)
plt.show()
```
