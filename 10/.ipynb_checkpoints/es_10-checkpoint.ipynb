# Lezione di laboratorio 10
In questa lezione è stato implementato in c++ un algoritmo SA (simulated annealing) per risolvere un problema TSP (Travelling Salesman Problem) come nel caso degli algoritmi genetici. <br>
Un Simulated Annealing è una tecnica euristica che si ispira alla "ricottura" dei metalli: questi vengono riscaldati e raffreddati lentamente allo scopo di eliminare gli errori nel reticolo cristallino. <br>
Questo metodo è utilizzato per trovare il minimo globale di una funzione in presenza di più minimi locali. <br>
In un sistema che si può trovre in molte e complesse configurazioni e in cui è possibile valutare ciscuna di queste con un'energia (o una funzione costo), partendo da soluzioni casuali si possono trovare le prossime soluzioni attraverso il metodo di Metropolis, e capionando come distribuzione: <br>
$$\    p=\left\{
    \begin{array}{c l}	
         e^{-\beta(L(x')-L(x))} & L(x')>L(x)\\
         1 & altrimenti
    \end{array}\right.$$<br>
Dove x è una configurazione e x' è un'altra possibile configurazione generata partendo da x (si è utilizzato un codice simile a quello per inserire mutazioni nelle nuove generazioni nell'algoritmo genetico) e L(x) è una funzione costo (nel caso TSP la lunghezza di un certo percorso). <br>
Invece $\beta$ rappresenta l'inverso della temperatura, quest'ultima viene abbassata lentamente. Questo metodo associa alle varie soluzioni un probabilità simile a quella di Boltzmann, in analogia con un sistema termodinamico. <br>

### Esercizio 10.1a
30 punti casuali sono generati su una circonferenza (r=1), e si cerca il percorso più breve che passa da tutti i punti una sola volta.
La simulazione si divide in 1000 step di temperatura decrescente, ciascuno dei quali campiona 1000 nuvoe configurazioni, "accettandole" o "rifiutandole" tramite Metropolis.
Di seguito sono riportati due grafici:
- funzione costo del miglior individuo trovato (in funzione del numero di step di temperatura) 
- miglior percorso trovato

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
#from matplotlib.pyplot import figure
a1,a2 = np.loadtxt("data/best_path_temp1.dat", usecols=(0,1), unpack=True)
plt.title("cost of the best individual")
plt.plot(a1,a2)
plt.show()
```

#### Miglior percorso trovato 

```python
import numpy as np
import matplotlib.pyplot as plt
cities_x, cities_y = np.genfromtxt("data/best_individual1.dat", usecols=(0,1), unpack=True)
plt.figure(figsize=(6, 6))
plt.plot(cities_x,cities_y,"-o")
#plt.quiver(citx[:-1], city[:-1], citx[1:]-citx[:-1],city[1:]-city[:-1])
plt.show()
```
