# Lezione di laboratorio 6
In questa lezione si è implementato un codice per simulare un modello di Ising in una dimensione.<br>
Un modello di Ising 1D è dato da una sequenza di particelle  che possono avere due valori di spin la cui hamiltoniana di interazione con quelle adiacenti è data da: <br>
$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$
Uno stato di questo sistema è quindi rappresentabile da una sequenza di +1 e -1. <br>
Sono stati implementati due metodi di campionamento del peso di Boltzmann $p(\{ s_i \}) = \exp(-\beta H)/Z$,il metodo di Metropolis e Gibbs.<br>

### Esercizio 6.1
Le simulazioni sono state svolte più volte a temperature diverse (a intervalli regolari tra 0.5 e 2.1) misurando ogni volta alcune proprietà del sistema:
<a>
1. Energia interna
2. Capacità termica
3. Suscettività magnetica
4. Magnetizzazione (h=0.02) <br> </a>
Di queste proprietà si è utilizzato l'ultimo valore ottenuto alla fine della simulazione (quello con incertezza minore) per farne un confronto con l'andamento teorico in funzione della temperatura.
La procedura è svolta sia per il campionamento di Gibbs che per quello di Metropolis.
Di seguito sono riportati in ogni grafico l'andamento teorico della proprietà indicata in funzione della temperatura e i risultati ottenuti con i due campionamenti.
Vi sono anche script in python (commentati) utilizzati per lanciare le varie simulazioni e raccogliere i risultati ottenuti.

```python
#python code for preparing output file for a confrontation with theoretical trend
#this may take 2 or 3 minutes
"""
DO NOT RUN THIS CODE

import os
import numpy as np
cmd1 = "./clean.sh"
cmd2 = "./Monte_Carlo_ISING_1D.exe"
T=0.5
metro = 1
h= 0.00
st = "sed -i '4c\{}' input.dat ".format(h)
os.system(st)
st = "sed -i '5c\{}' input.dat ".format(metro)
os.system(st)
os.system("rm metro*")
os.system("rm gibbs*")
e=open("metro.ene.txt", "a+")
c=open("metro.heat.txt", "a+")
x=open("metro.chi.txt", "a+")
eg=open("gibbs.ene.txt", "a+")
cg=open("gibbs.heat.txt", "a+")
xg=open("gibbs.chi.txt", "a+")
while T <=2.1:
    os.system(cmd1)
    st = "sed -i '1c\{}' input.dat ".format(T)
    os.system(st)
    os.system(cmd2)
    e1, e2 = np.loadtxt("output.ene.0", usecols=(2, 3), unpack=True)
    c1, c2 = np.loadtxt("output.heat.0", usecols=(2, 3), unpack=True)
    x1, x2 = np.loadtxt("output.chi.0", usecols=(2, 3), unpack=True)
    st2 = "{} {} {}\n".format(T,e1[len(e1)-1],e2[len(e2)-1])
    e.write(st2)
    st2 = "{} {} {}\n".format(T,c1[len(c1)-1],c2[len(c2)-1])
    c.write(st2)
    st2 = "{} {} {}\n".format(T,x1[len(x1)-1],x2[len(x2)-1])
    x.write(st2)
    T+=0.1
e.close()
c.close()
x.close()
#for Gibbs sampling
#wait up to 3 minutes again
metro = 3
st = "sed -i '5c\{}' input.dat ".format(metro)
os.system(st)
T=0.5
while T <=2.1:
    os.system(cmd1)
    st = "sed -i '1c\{}' input.dat ".format(T)
    os.system(st)
    os.system(cmd2)
    e1, e2 = np.loadtxt("output.ene.0", usecols=(2, 3), unpack=True)
    c1, c2 = np.loadtxt("output.heat.0", usecols=(2, 3), unpack=True)
    x1, x2 = np.loadtxt("output.chi.0", usecols=(2, 3), unpack=True)
    st2 = "{} {} {}\n".format(T,e1[len(e1)-1],e2[len(e2)-1])
    eg.write(st2)
    st2 = "{} {} {}\n".format(T,c1[len(c1)-1],c2[len(c2)-1])
    cg.write(st2)
    st2 = "{} {} {}\n".format(T,x1[len(x1)-1],x2[len(x2)-1])
    xg.write(st2)
    T+=0.1
eg.close()
cg.close()
xg.close()
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
points=100
T = np.linspace(0.2,3.0,num=points)
m1,m2,m3 = np.loadtxt("data/metro.ene.txt", usecols=(0,1,2), unpack=True)
g1,g2,g3 =np.loadtxt("data/gibbs.ene.txt", usecols=(0,1,2), unpack=True)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
s=1.5
plt.figure(figsize=(6.4*s, 4.8*s))
plt.plot(T, e)
plt.title('Ising 1D, internal energy (20 blocks, 10000 steps each)')
plt.xlabel('T')
plt.ylabel('U/N')
plt.errorbar(m1, m2, yerr=m3)
plt.errorbar(g1,g2,yerr=g3)
plt.legend(["theoretical trend", "Metropolis sampling", "Gibbs sampling"])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
m1,m2,m3 = np.loadtxt("data/metro.heat.txt", usecols=(0,1,2), unpack=True)
g1,g2,g3 =np.loadtxt("data/gibbs.heat.txt", usecols=(0,1,2), unpack=True)
s=1.5
plt.figure(figsize=(6.4*s, 4.8*s))
plt.plot(T, heat)
plt.title('Ising 1D, Heat capacity')
plt.xlabel('T')
plt.ylabel('C')
plt.errorbar(m1, m2, yerr=m3)
plt.errorbar(g1,g2, yerr=g3)
plt.legend(["theoretical trend", "Metropolis sampling", "Gibbs sampling"])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
m1,m2,m3 = np.loadtxt("data/metro.chi.txt", usecols=(0,1,2), unpack=True)
g1,g2,g3 = np.loadtxt("data/gibbs.chi.txt", usecols=(0,1,2), unpack=True)
s=1.5
plt.figure(figsize=(6.4*s, 4.8*s))
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(T, X)
plt.title('Ising 1D, Susceptibility')
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.errorbar(m1, m2, yerr=m3)
plt.errorbar(g1,g2,yerr=g3)
plt.legend(["theoretical trend", "Metropolis sampling", "Gibbs sampling"])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
m1,m2,m3 = np.loadtxt("data/metro.mag.txt", usecols=(0,1,2), unpack=True)
g1,g2,g3 = np.loadtxt("data/gibbs.mag.txt", usecols=(0,1,2), unpack=True)
s=1.5
plt.figure(figsize=(6.4*s, 4.8*s))
T = np.linspace(0.2,3.0,num=points)
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(T, M)
plt.title('Ising 1D, magnetization M with h = 0.02')
plt.xlabel('T')
plt.ylabel('$M$')
plt.errorbar(m1, m2, yerr=m3)
plt.errorbar(g1,g2,yerr=g3)
plt.legend(["theoretical trend", "Metropolis sampling", "Gibbs sampling"])
plt.show()
```
