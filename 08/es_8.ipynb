# Lezione di laboratorio 8
In questa lezione si è implementato un codice c++ per trovare il ground state di un sistema quantistico di una particella in una dimensione soggetta ad un potenziale:
$$
V(x) = x^4 - \frac{5}{2}x^2
$$
Il metodo utilizzato è variazionale, ovvero non conoscendo la soluzione analitica di questo problema si ipotizza che la funzione d'onda che lo risolve sia:
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$
e si variano i parametri $\mu$ e $\sigma$ per cercare il valore di aspettazione minimo dell'energia.

Il seguente istogramma mostra la frequenza delle posizioni campionate dal programma, confrontate con la funzione d'onda (in modulo) trovata risolvendo numericamente l'equazione di schroedinger

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math as mt
samp = np.loadtxt("data/samples.dat", usecols=(0), unpack=True)
fig = plt.figure(figsize=(12.8,9.6))
plt.hist(samp, bins=100, range=(-2, 2), density=True, stacked=True)

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
scale = 0.3
plt.plot(x,((psi[0])**2))
plt.xlabel("x")
plt.grid(True)
plt.xlim((-2,2))
plt.show()
```
